```python
def Titanic_clean(data_type, path0): #test or train as string
    file = os.path.join(path0, data_type + ".csv")
    df = pd.read_csv(file)  
    df.replace('female', 1, inplace = True)
    df.replace('male', 0, inplace = True)
    df['Embarked'].replace('S', 0, inplace = True)
    df['Embarked'].replace('C', 1, inplace = True)
    df['Embarked'].replace('Q', 2, inplace = True)
    Id = df['PassengerId']
    df.drop(['PassengerId','Name', 'Ticket', 'Cabin'], axis = 1, inplace = True)
    df.fillna(method='ffill', inplace = True)
    if data_type == 'train':
        y = df['Survived']
        X = df.drop('Survived', axis = 1)
        return y,X,Id
    else:
        X = df
        return X, Id
    
```

```python
path0 = r'C:\Users\mvariani\Desktop\Projects\Courses\Kaggle\Titanic'.replace('\\', '/')
y_train, X_train = Titanic_clean('train',path0)
X_test = Titanic_clean('test',path0)
X_test.describe()
```

```python
y_train, X_train, Id = Titanic_clean('train',path0)
X_test, Id = Titanic_clean('test',path0)

from sklearn.svm import SVC
model_svc = SVC(C = 1)
model_svc.fit(X_train, y_train)
y_pred_svc = model_svc.predict(X_test)

PassengerId = Id
Survived = y_pred_svc
print(PassengerId, Survived)
```
